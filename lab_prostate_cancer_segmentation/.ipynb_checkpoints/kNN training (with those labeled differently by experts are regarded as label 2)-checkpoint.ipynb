```python
from kNN_function_with_kFold import kNN_k_parameter_adjust
from kNN_function_with_kFold import kNN_data_ratio_adjust
#from kNN_function import kNN_data_ratio_adjust_2
from kNN_function_with_kFold import kNN_k_parameter_adjust_with_specific_data_ratio 
from plot_prediction import kNN_plot_prediction
import matplotlib.pyplot as plt
```

```python
X_train = X_after_CNN
y_train = y_after_CNN
```

```python
label_list = y_train ['label']

label_list=label_list.replace([1.5],[2])

y_train ['label'] = label_list
```

```python
k_value = 15
kNN_scaled_CNN_result_data_ratio=kNN_data_ratio_adjust(X_train,y_train,k_value)
```

```python
plt.title("kNN ROC with data ratio changing (k=15) - cross validation stage")
plt.plot(kNN_scaled_CNN_result_data_ratio['FPR'],kNN_scaled_CNN_result_data_ratio['TPR'],"r-",label='scaled_CNN_data')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

```python
plt.title("kNN f1_score with data ratio changing (k=15) - cross validation stage")
plt.plot(kNN_scaled_CNN_result_data_ratio['label 1: label 2 ratio'],kNN_scaled_CNN_result_data_ratio['f1_score'],"r-",label='scaled_CNN_data')
plt.xlabel('label 1: label 2 ratio')
plt.ylabel('f1_score')
plt.legend()
plt.show()
```

```python
plt.title("kNN BER with data ratio changing (k=20) - cross validation stage")
plt.plot(kNN_scaled_CNN_result_data_ratio['label 1: label 2 ratio'],kNN_scaled_CNN_result_data_ratio['BER'],"r-",label='scaled_CNN_data')
plt.ylabel('BER')
plt.legend()
plt.show()
```

```python
kNN_scaled_CNN_result_k_value_specific_1_ratio =kNN_k_parameter_adjust_with_specific_data_ratio (X_train,y_train,1)
kNN_scaled_CNN_result_k_value_specific_11_ratio =kNN_k_parameter_adjust_with_specific_data_ratio (X_train,y_train,1.1)
kNN_scaled_CNN_result_k_value_specific_12_ratio =kNN_k_parameter_adjust_with_specific_data_ratio (X_train,y_train,1.2)
kNN_scaled_CNN_result_k_value_specific_13_ratio =kNN_k_parameter_adjust_with_specific_data_ratio (X_train,y_train,1.3)
kNN_scaled_CNN_result_k_value_specific_14_ratio =kNN_k_parameter_adjust_with_specific_data_ratio (X_train,y_train,1.4)
kNN_scaled_CNN_result_k_value_specific_15_ratio =kNN_k_parameter_adjust_with_specific_data_ratio (X_train,y_train,1.5)
```

```python
plt.title("kNN f1_score with k changing - cross validation stage")
plt.plot(kNN_scaled_CNN_result_k_value_specific_1_ratio['k_value'],kNN_scaled_CNN_result_k_value_specific_1_ratio['f1_score'],"b-",label='scaled_CNN_data (data ratio:1)')
plt.plot(kNN_scaled_CNN_result_k_value_specific_11_ratio['k_value'],kNN_scaled_CNN_result_k_value_specific_11_ratio['f1_score'],"y-",label='scaled_CNN_data (data ratio:1.1)')
plt.plot(kNN_scaled_CNN_result_k_value_specific_12_ratio['k_value'],kNN_scaled_CNN_result_k_value_specific_12_ratio['f1_score'],"g-",label='scaled_CNN_data (data ratio:1.2)')
plt.plot(kNN_scaled_CNN_result_k_value_specific_13_ratio['k_value'],kNN_scaled_CNN_result_k_value_specific_13_ratio['f1_score'],"r-",label='scaled_CNN_data (data ratio:1.3)')
plt.plot(kNN_scaled_CNN_result_k_value_specific_14_ratio['k_value'],kNN_scaled_CNN_result_k_value_specific_14_ratio['f1_score'],"k-",label='scaled_CNN_data (data ratio:1.4)')
plt.plot(kNN_scaled_CNN_result_k_value_specific_15_ratio['k_value'],kNN_scaled_CNN_result_k_value_specific_15_ratio['f1_score'],"m-",label='scaled_CNN_data (data ratio:1.5)')
plt.xlabel('k_value')
plt.ylabel('f1_score')
plt.legend()
plt.show()
```
